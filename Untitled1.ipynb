```python
import pandas as pd

#Taxonomy spreadsheet -> extract all IDs related to lentivirus
taxonomy_keys = pd.read_csv('taxonomy_keys_10239.csv')
print(taxonomy_keys.shape)
taxonomy_keys.head()
```

```python
#extract all IDs related to lentivirus
tmp = taxonomy_keys.iloc[8]['IDs']

#remove commas + whitespace + create list
lentivirusID_list = lentivirusID = [id.strip() for id in tmp.split(',')]

#turn string to int
lentivirusID = [int(x) for x in lentivirusID_list]
```

```python
#Open big dataset -> isolate all rows having lentivirusIDs and save to seperate file
data = pd.read_csv('all_virus_LBCE_with_ESM1b_features.csv')
data.head()
```

```python
#isolate all rows having lentivirusID
lentivirus_data = data[data['Info_organism_id'].isin(lentivirusID)]
#save to seperate file 'lentivirus'
lentivirus_data.to_csv('lentivirus.csv', index=False)
```

```python
#load lentivus data
lentivirus = pd.read_csv('lentivirus.csv')
#check the size of data
lentivirus.shape
lentivirus.head()
```

```python
#check for missing values -> results show no missing values
print("Total Missing:")
for x, y in enumerate(lentivirus.isnull().any()):
    if y == True:
        name = lentivirus.columns[x]
        print("   "+name+":", data[name].isnull().sum())
```

```python
from matplotlib import pyplot as plt
#check class balance
print("Class Balance:")
print("   Positive:",len(lentivirus[lentivirus["Class"]==1]))
print("   Negative:",len(lentivirus[lentivirus["Class"]==-1]))   

#visualize class imbalance
class_col = lentivirus['Class']
class_counts = class_col.value_counts()
plt.bar(class_counts.index, class_counts.values)

plt.xlabel('Class')
plt.title('Counts of Class Values')

# Show the plot
plt.show()
```

```python
#check + visualize clusters
lentivirus_clusters = lentivirus.copy(deep=True)
#lentivirus_clusters = lentivirus_clusters.iloc[:, 12:]

counts = lentivirus_clusters.groupby(['Info_cluster', 'Class']).size().unstack()
counts.plot(kind='bar', stacked=True)
print(counts)
plt.title('Positive and Negative Samples in Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.model_selection import GroupKFold
#split data into 80-20 split
X = lentivirus_clusters.drop('Class', axis=1)
y = lentivirus_clusters['Class']
groups = lentivirus_clusters['Info_cluster']

gkfold = GroupKFold(n_splits=3)

for train_idx, test_idx in gkfold.split(X, y, groups):
    X_train, y_train= X.loc[train_idx], y.loc[train_idx]
    X_test, y_test= X.loc[test_idx], y.loc[test_idx]
```

```python
X2 = X_train.reset_index(drop=True)
y2 = y_train.reset_index(drop=True)
groups2 = X_train['Info_cluster']

gkfold = GroupKFold(n_splits=2)

for train_idx, test_idx in gkfold.split(X2, y2, groups2):
    X2_train, y2_train= X2.loc[train_idx], y2.loc[train_idx]
    X_val, y_val= X.loc[test_idx], y.loc[test_idx]

X2.shape
```
