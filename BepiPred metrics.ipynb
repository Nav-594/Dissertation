```python
import pandas as pd

holdout = pd.read_csv('holdout_data.csv')
print(holdout.shape)
holdout.head()
```

These are the prediction gotten from BepiPred 3.0

```python
fasta_data = """
>ABI20203.1
krwiilglnkivrmYGPKEPfrdyvdrfyk
>AAC03762.1
reitfiykssctdsdhcqeyqckkvnLNsSDSERQQVEetfnligcierthvfchtg
>P16082.1
snsvrvedvtNTaEYwgfk
>P23426.1
tgkipwILlpgr
"""  
fasta_data
```

Making the Fasta formatted data into workable format

```python
# Parse FASTA data into a dictionary
fasta_dict = {}
lines = fasta_data.strip().split('\n')
current_id = None
current_sequence = []
for line in lines:
    if line.startswith(">"):
        if current_id is not None:
            fasta_dict[current_id] = ''.join(current_sequence)
        current_id = line[1:]
        current_sequence = []
    else:
        current_sequence.append(line.strip())
fasta_dict[current_id] = ''.join(current_sequence)

# Get predicted epitopes from FASTA data
predicted_epitopes = {protein_id: sequence for protein_id, sequence in fasta_dict.items() if any(char.isupper() for char in sequence)}

predicted_epitopes
```

Making a dataset using the BepiPred 3.0 predictions, capital letters showcased a epitope so labelled as 1 and vice versa for non-epitopes

```python
# Create an empty list to store DataFrame chunks
data_frames = []

# Loop through fasta_dict and populate the DataFrame
for protein_id, sequence in fasta_dict.items():
    # Create a DataFrame for the current protein sequence
    df_chunk = pd.DataFrame({
        'Info_protein_id': [protein_id] * len(sequence),
        'Info_AA': list(sequence),
        'Class': [1 if aa.isupper() else -1 for aa in sequence]
    })
    # Append the DataFrame chunk to the list
    data_frames.append(df_chunk)

# Concatenate all DataFrame chunks into a single DataFrame
epitope_data = pd.concat(data_frames, ignore_index=True)
```

Made the dataset with BepiPred 3.0 predictions

```python
# Specify the column name
column_name = 'Info_protein_id'

# Get unique values in the specified column
unique_values = epitope_data[column_name].unique()

# Print unique values
print(unique_values)
print(epitope_data.shape)
```

For comparison, get the holdout set and make it the same as the BeciPred 3.0 prediction dataset. So we can use performance metrics on BepiPred 3.0 prediction

```python
#olumns_to_keep = ['Info_protein_id','Info_AA', 'Class']
validation = holdout[holdout['Info_protein_id'].isin(unique_values)]
validation
```

```python
keep = {'Info_protein_id','Info_AA','Class'}
validation_copy = validation.copy()
validation_copy.drop(validation.columns.difference(keep), axis=1, inplace=True)
validation_copy
```

```python
validation_copy.to_csv('validation.csv', index=False)
```

```python
true_labels = validation_copy["Class"]
predicted_labels = epitope_data["Class"]
```

Performance metrics used on BeciPred 3.0 data

```python
import seaborn as sns
from sklearn.metrics import f1_score, accuracy_score, classification_report, confusion_matrix, roc_auc_score, precision_score, recall_score, matthews_corrcoef

auc_score = roc_auc_score(true_labels, predicted_labels)
mcc = matthews_corrcoef(true_labels, predicted_labels)
sns.heatmap(confusion_matrix(true_labels, predicted_labels), annot = True)
print(classification_report(true_labels, predicted_labels))
print(f"Accuracy: {round(accuracy_score(true_labels, predicted_labels), 2)}") 
print(f"Precision: {round(precision_score(true_labels, predicted_labels), 2)}")
print(f"Recall: {round(recall_score(true_labels, predicted_labels), 2)}")
print(f"F1_score: {round(f1_score(true_labels, predicted_labels), 2)}")
print("AUC Score:", auc_score)
print("Matthews Correlation Coefficient (MCC):", mcc)
```
