```python
import pandas as pd

test_data = pd.read_csv('test_data.csv')
print(test_data.shape)
test_data.head()
```

```python
unique_protein_ids = test_data['Info_protein_id'].unique()
print(unique_protein_ids)
```

```python
grouped = test_data.groupby('Info_protein_id')

# Iterate through unique Info_protein_id values and their respective Info_AA and Info_pos
for protein_id, group_data in grouped:
    unique_aa_pos = group_data[['Info_AA', 'Info_pos']].drop_duplicates()
    
    print(f"Unique Info_protein_id: {protein_id}")
    print(unique_aa_pos)
```

```python
for protein_id, group_data in grouped:
    sorted_group = group_data.sort_values(by='Info_pos')
    
    print(f"Unique Info_protein_id: {protein_id}")
    print(sorted_group[['Info_AA', 'Info_pos']])
```

```python
# Create an empty dictionary to store sequences for each protein_id
sequences = {}

# Iterate through unique Info_protein_id values and their respective Info_AA and Info_pos
for protein_id, group_data in grouped:
    # Sort the group_data by Info_pos
    sorted_group = group_data.sort_values(by='Info_pos')
    
    # Extract Info_AA column as the sequence
    sequence = ''.join(sorted_group['Info_AA'])
    
    # Store the sequence in the dictionary with protein_id as the key
    sequences[protein_id] = sequence

# Generate FASTA format
fasta_lines = []
for protein_id, sequence in sequences.items():
    fasta_lines.append(f">{protein_id}")
    fasta_lines.append(sequence)

# Write the FASTA format to a file or print it
with open('output.fasta', 'w') as fasta_file:
    fasta_file.write('\n'.join(fasta_lines))

# Print the FASTA format
print('\n'.join(fasta_lines))
```
